##Scatter Plot

```python
x=np.arange(0,10,1)
y=np.array((1,2,3,2.8,2.4,2,2.3,2.7,2.3,2))
plt.scatter(x,y)
```

```python
x=np.array((1,2,1,2,5,3,4,5,7,8,6,7,8,3,2,1))
plt.hist(x)
```

```python
t=np.arange(-10,10,0.1)
fun=np.sin(t)
xlabel("t")
ylabel("$\sin(t)$")
plot(t,fun,color='red',linestyle='dotted',linewidth='2',label='$\sin(x)$ Function')  #dashed
```

#Three-Dimensional Plotting

```python
'''
The 3d plots are enabled by importing the mplot3d toolkit
'''
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
z=np.linspace(-6,6,100)
x=np.sin(2*z)
y=np.cos(2*z)
# ax.plot3D(x,y,z)
ax.plot3D(x,y,z)
ax.scatter(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
from mpl_toolkits import mplot3d
ax.set_title('3D line plot')
fig = plt.figure()
ax = plt.axes(projection='3d')
z=np.linspace(-6,6,100)
x=np.sin(2*z)
y=np.cos(2*z)
# ax.plot3D(x,y,z)
ax.plot3D(x,y,z)
ax.scatter(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

#Three-dimensional Contour Plots

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 60, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
'''
an elevation of 60 degrees (that is, 60 degrees above the x-y plane) 
and an azimuth of 35 degrees (that is, rotated 35 degrees counter-clockwise about the z-axis):
'''
ax.view_init(60, 35)
fig
```
